```python
def get_process_json(data_path,to_path):
    data = pd.read_csv(data_path,encoding='utf-8')
    try:
        data['label'].loc[data['label']==-1] = 0
    except:
        print(1)
    data_all = []
    no_use_word = [',','.',',','"','|','"','《','》','，','。','？','/','、',';','；','：',':','!','！','的','等','了','嗯','那']
    for i in range(data.shape[0]):
        temp_dict = dict()
        temp_dict['time'] = str(data.loc[i,'date'])
        temp_dict['label'] = int(data.loc[i,'label'])
        sent = str(data.loc[i,'sent'])
        temp_dict['sent']  = sent
        words = jieba.lcut(sent)
        cut_words = pd.Series(words)
        res_words = cut_words[~cut_words.isin(no_use_word)].tolist()
        temp_dict['words'] = res_words
        data_all.append(temp_dict)
    with open(to_path,'w') as f:
        json.dump(data_all,f,ensure_ascii=False)
    return data_all
get_process_json('../data/data1.csv','../data/data1.json')
```

```python
def combine_data():
    with open('../data/data1.json','r') as f:
        data1 = json.load(f)
    with open('../data/data2.json','r') as f:
        data2 = json.load(f)
    data1 += data2
    with open('../data/data.json','w') as f:
        json.dump(data1,f,ensure_ascii=False)
combine_data()
```

```python
def get_word_to_ix(data_path):
    with open(data_path,'r') as f:
        data = json.load(f)
    word_all = []
    for i in data:
        word_all += i['words']
    res = dict()
    i = 0
    for word in word_all:
        if word not in res.keys():
            res[word] = i
            i+=1
    with open('../data/word_to_ix.json','w') as f:
        json.dump(res,f,ensure_ascii=False)
get_word_to_ix('../data/data.json')
```
