```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd
import numpy as np
```

```python
def get_data(file_path):
    data = pd.read_excel(file_path)
    return data
data = get_data('price_data2.xlsx')
data
```

```python
price_list = list(data['柑橘价格'])
price_change = [0]
for i in range(len(price_list)-1):
    price_change.append(price_list[i+1] - price_list[i])
price_change
```

```python
data2 = data.copy()
data2.insert(loc=3,column='价格波动',value=0)
for i in range(data2.shape[0]):
    data2.loc[i,'价格波动'] = price_change[i]
data2
```

```python
data2.to_excel('granger_price_prob.xlsx',index=None)
```
