```python
import pandas as pd
import json 
import numpy as np
```

```python
def get_data(file_path):
    with open(file_path,'r') as f:
        data = json.load(f)
    return data
```

```python
def get_price(file_path):
    with open(file_path,'r') as f:
        data = json.load(f)
    data = data['datas']
    for i in data:
        i['DATA'] = float(i['DATA'])
        i['DATADATE'] = i['DATADATE'].replace('-','/')
    return  data
data = get_price('price.json')
```

```python
def to_csv(data):
    pd_data = pd.DataFrame(data)
    pd_data = pd_data.drop(columns=['NAME','UNIT'])
    pd_data = pd_data.loc[:,['DATADATE','DATA']]
    pd_data.to_csv('price_res.csv',index=None)
    return pd_data
price_data = to_csv(data)
```

```python
pred_data = get_data('pred_res.json')
pred_data
```

```python
def change_time(data):
    for i in data:
        time:str = i['time']
        time = time.split(" ")[0].split('/')
        for k in range(len(time)):
            time[k] = time[k].rjust(2,'0')
        time = '/'.join(time)
        i['time'] = time
        # i['time'] = i['time'].split(" ")[0]
    with open('pred_res2.json','w') as f:
        json.dump(data,f,ensure_ascii=False)
    return True
change_time(pred_data)
```

```python
def time_check(data,time):
    res = []
    for i in data:
        if i['time'] == time:
            res.append(i['pred_prob'][0])
    return res
a = time_check(pred_data,'2020/10/18')
if len(a) > 0:
    print(np.array(a).mean())
else:
    print('no')
```

```python

def data_para(_price_data:pd.DataFrame,pred_data:pd.DataFrame):
    price_data = _price_data.copy()
    price_data.insert(loc=2,column='PROB',value=-1)
    for i in range(price_data.shape[0]):
        time = str(price_data.loc[i,'DATADATE'])
        print(time)
        prob_list = time_check(pred_data,time)
        if len(prob_list) > 0:
            prob = np.array(prob_list).mean()
            price_data.loc[i,'PROB'] = prob
    return price_data
price_data2 = data_para(price_data,pred_data)
price_data2
```
